```python
import pandas as pd
```

```python
data=pd.read_csv('Water Quality Testing.csv')
data.sample()
```

FILLING MISSING VALUE WITH 0

```python
data_zero=data.fillna(0)
```

```python
data_zero.isnull().sum()
```

filling missing data with mean value

```python
data['Turbidity_update (NTU)']=data['Turbidity (NTU)'].fillna(data['Turbidity (NTU)'].mean())
data['Dissolved Oxygen_update (mg/L)']=data['Dissolved Oxygen (mg/L)'].fillna(data['Dissolved Oxygen (mg/L)'].mean())
data['Conductivity_update (µS/cm)']=data['Conductivity (µS/cm)'].fillna(data['Conductivity (µS/cm)'].mean())
data
```

APPLY ANY OPERATION TO ROW

```python
data['Temperature_update (°C)']=data['Temperature (°C)'].apply(lambda x:x+1)
data.head()
```
