```python
import matplotlib.pyplot as plt
```

LINEPLOT

```python
y=[3,2,4,5,7,9]
x=[1,2,3,4,5,6]

plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('random graph')
plt.show()
```

```python
y=[3,2,4,5,7,9]
x=[1,2,3,4,5,6]

plt.plot(x,y,color='red',linestyle='-.',marker='o',markersize=10,linewidth=3)
```

```python
y=[3,2,4,5,7,9]
x=[1,2,3,4,5,6]
z=[1,3,6,8,3,7]

plt.figure(figsize=(8,4))

plt.subplot(2,2,1)
plt.plot(x,y,color='red')
plt.title('plot 1')

plt.subplot(2,2,2)
plt.plot(x,y,color='green')
plt.title('plot 2')

plt.subplot(2,2,3)
plt.plot(x,y,color='pink')
plt.title('plot 3')
```

BARPLOT

```python
x=[1,2,3,4,5]
y=[1,3,7,13,25]

plt.bar(x,y)
```

```python
x=[1,1,3,3,4,5,5,5,5]

plt.hist(x)
```

```python
x=[1,2,3,4,5]
y=[1,3,7,13,25]

plt.scatter(x,y,color='red',marker='x')
```
