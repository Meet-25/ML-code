IMPORT PANDAS

```python
import pandas as pd
```

DATA TYPE = series

pandas series is a 1D array like object which can hold ANY DATA

```python
## using list
lst=['a','b','c','d','e']
data_lst=pd.Series(lst)
print(data_lst)
type(data_lst)
```

```python
## using dictionary
dict={1:'a',2:'b',3:'c'}
data_dict=pd.Series(dict)
print(data_dict)
type(data_dict)
```

```python
lst1=[1,2,3,4]
lst2=['a','b','c','d']
data1=pd.Series(lst1,index=lst2)
print(data1)
```

DATA TYPE = dataframe

dataframe is a multi dimensional data which also can hold any data 

```python
##creating daqtaframe from dictionary of list
dict={
    "Name":['Meet','ram','shiv'],
    "Age":[20,22,26],
    "income":[5000000,3000000,5000000]  
}

data2=pd.DataFrame(dict)
print(data2)
type(data2)
```

```python
##creating daqtaframe from list of dictionary
dict2=[
    {'Name':'meet','Age':20,'income':5000000},
    {'Name':'Mavani','Age':23,'income':3000000},
    {'Name':'jiyan','Age':25,'income':5000000}  
]

data3=pd.DataFrame(dict2)
print(data3)
type(data3)
```
